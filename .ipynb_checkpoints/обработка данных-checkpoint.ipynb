```python
# Функция считывания наименования исходных датасетов
def list_name():
    lst_name = list()
    for filename in os.listdir('..\.\Данные'):
        if filename.endswith('.xlsx'):
            lst_name.append("..\.\Данные\\" + filename)
    return lst_name
```

```python
# Фукнция считывания, первичной обработки и объединения датасетов
def form_dataset(lst: list):
    res = pd.DataFrame()
    for i in lst:
        df_old = pd.read_excel(i)
        
        df_old = df_old[col]

        df_old = df_old[(df_old['Регион']=='Екатеринбург') | 
                        (df_old['Регион']=='Березовский') | 
                        (df_old['Регион']=='Верхняя Пышма') | 
                        (df_old['Регион']=='Ревда') | 
                        (df_old['Регион']=='Каменск-Уральский') | 
                        (df_old['Регион']=='Нижний Тагил') | 
                        (df_old['Регион']=='Другие населенные пункты')]


        df_old = df_old.rename(columns={'Взвешенная медианная стоимость кв.м. первичка 2-ком.': f'mn_pc_new_{i[37:44]}',
                                        'Взвешенная медианная стоимость кв.м. вторичка 2-ком.': f'mn_pc_old_{i[37:44]}',
                                        'Количество оферов продажа первичка 2-ком.': f'num_sale_cnt_new_{i[37:44]}',
                                        'Количество оферов продажа вторичка 2-ком.': f'num_sale_cnt_old_{i[37:44]}',
                                        'Регион': 'region',
                                        'Средний срок экспозиции, вторичка': f'period_sale_{i[37:44]}'})
        
        if len(res) == 0:
            res = df_old.copy()
        else:
            res = pd.merge(res, df_old, on='region')
    return res    
```

```python
# Считывание списка наименований исходных данных
lst_data = list_name()
lst_data
```

```python
# функция конвертации строковых типов данных в числовой
def convert_value(data, lst):
    for town in lst:
        data[town] = data[town].str.replace('₽', '', regex=True).replace(' ', '', regex=True)
        data[town] = pd.to_numeric(data[town])
    return data
```

```python
# применим функцию для конвертации значений в датасете
price_new_apartments = convert_value(price_new_apartments, col_town)
price_new_apartments
```

```python
# применим функцию для конвертации значений в датасете
count_new_apartments = convert_value(count_new_apartments, col_town)
count_new_apartments
```

```python
# напишем функцию, которая будет заполнять пустые значения в строках медианным значением
def func_fillna_median(data):
    for col in data:
        if data[col].isna().sum() > 0:
            data[col].fillna(data[col].median(), inplace=True)
    return data
```

```python
# напишем функцию, которая будет заполнять пустые значения 0
def func_fillna_zero(data):
    for col in data:
        if data[col].isna().sum() > 0:
            data.loc[data[col].isnull(), col] = 0
    return data
```

```python
# Применим функцию заполнения пустых значений медианным значением
price_new_apartments = func_fillna_median(price_new_apartments)
price_new_apartments
```

```python
# Применим функцию заполнения пустых значений нулевым значением
count_new_apartments = func_fillna_zero(count_new_apartments)
count_new_apartments
```

```python
# Напишем функцию конвертации дней в часы и переведем данные из строкового формата в числовой
def conver_to_hour(data, lst):
    for col in lst:
        data[col] = data[col].apply(lambda x: int(x.split()[0]) * 24 + int(x.split()[2]) 
                                                                            if len(x.split()) > 2
                                                                            else int(x.split()[0]) * 24)
    return data
```

```python
# Применим функцию конвертации дней в часы на датафрейме
mean_period_price = conver_to_hour(mean_period_price, col_town)
mean_period_price
```

Применим тест Шапиро - Уилка

```python
alpha = 0.05
norm_dist_price_new = list()
not_norm_dist_price_new = list()

for town in price_new_apartments[col_town]:
    
    stat, p_value = shapiro(price_new_apartments[town])
    
    if p_value > alpha:
        print('Нормальное распределение (не отвергаем H0)')
        norm_dist_price_new.append(town)
    else:
        print('Не нормальное распределения (отвергаем H0)')
        not_norm_dist_price_new.append(town)
    
    print(f"Город: {town}")
    print(f"Шапиро-Уилк тест: статистика = {stat: .4f}, p-значение = {p_value: .4f}")
    print("-"*50)
```

```python
norm_dist_price_old = list()
not_norm_dist_price_old = list()

from scipy.stats import shapiro

for town in price_old_apartments[col_town]:
    
    stat, p_value = shapiro(price_old_apartments[town])
    
    if p_value > alpha:
        print('Нормальное распределение (не отвергаем H0)')
        norm_dist_price_old.append(town)
    else:
        print('Не нормальное распределения (отвергаем H0)')
        not_norm_dist_price_old.append(town)
    
    print(f"Город: {town}")
    print(f"Шапиро-Уилк тест: статистика = {stat: .4f}, p-значение = {p_value: .4f}")
    print("-"*50)
```

```python
# найдем пересечение 2-х список городов с нормальным распределением

interseption_town_price = [town for town in norm_dist_price_new if town in norm_dist_price_old]
interseption_town_price
```

```python
# найдем пересечение 2-х список городов с ненормальным распределением

interseption_town_price_not = [town for town in not_norm_dist_price_new if town in not_norm_dist_price_old]
interseption_town_price_not
```

```python
# Т.к. тест Шапиро - Уилка показал нам, что есть признаки распределенные нормально, поэтому применяем тест Стюдента
# для этих признаков

for town in interseption_town_price:
    t_value, p_value = stats.ttest_ind(price_new_apartments[town], price_old_apartments[town])
    
    print(f"Город: {town}")
    print(t_value, p_value)

    if p_value > alpha: # Интерпретируем
        print('Одинаковые распределения (не отвергаем H0)')
        print('-'*50)
    else:
        print('Разные распределения (отвергаем H0)')
        print('-'*50)
    
```

```python
# Т.к. в выборке у нас есть признаки распределеные не нормально, при этом у нас независимые данные
# то мы применяем тест Манна-Уитни

for town in interseption_town_price_not:
    
    t_value, p_value = mannwhitneyu(price_new_apartments[town],  price_old_apartments[town])
    
    print(f"Город: {town}")
    print(f'Статистики: {t_value}, p_value = {p_value}')
    print()

    if p_value > alpha:
        print("Принимаем нулевую гипотезу, т.е. Н0")
    else:
        print("Принимаем альтернативную гипотезу, т.е. Н1")
```
